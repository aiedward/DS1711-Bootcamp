```python
df = pd.read_csv('/Users/panwu/Downloads/abalone.data', header=None)
```

```python
df.columns = ['Sex', 'Length', 'Diameter', 'Height', 
              'Whole weight', 'Shucked weight', 'Viscera weight', 
              'Shell weight', 'Rings']
```

```python
df.Rings.value_counts().sort_index().plot.bar()
```

```python
data = df.Rings.value_counts().sort_index()
```

```python
fig, ax = plt.subplots(1, 1, figsize=[10, 6])
ax.bar(data.index, data.values)
ax.set_title('Abacone age distribution', fontsize=20)
# ax.set_xlabel('Age', fontsize=20)
ax.set_ylabel('Count', fontsize=20)
ax.xaxis.set_label_text('Age', fontsize=20)
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=20)

# ax.text(15, 400, 'Highest point around 10', fontsize=20)
ax.annotate(xy=(data.idxmax(), data.max()), xytext=(data.idxmax()+2, data.max()-100), 
            s='Highest point around 10', fontsize=20, 
            arrowprops={'width': 0.2})
```

```python
df.plot.scatter(x='Length', y='Rings')
```

```python
fig, ax = plt.subplots(1, 1, figsize=[8, 8])
ax.hexbin(data=df, x='Length', y='Rings', gridsize=25, cmap='Blues')
ax.set_ylabel('Rings', fontsize=20)
ax.xaxis.set_label_text('Length', fontsize=20)
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=20)
```
